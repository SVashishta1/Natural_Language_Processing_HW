```python
import spacy
from spacy.lang.en.stop_words import STOP_WORDS
import string
from heapq import nlargest



with open('spaceX_DP.txt', 'r') as file: # reading the file content
    text = file.read()

print(text[:100]) # checking the data
```

```python
print(string.punctuation, STOP_WORDS)
```

```python
nlp = spacy.load('en_core_web_sm')  # loading the English model
doc = nlp(text)
unwanted_chars = string.punctuation + "“”‘’—"  # Adding smart quotes and em dash

tokens = []

# tokenizing the text and remove stopwords & punctuations

for w in doc:
# checking for a stopword, punctuation, and allowing only alphabetic characters
#because I have seen numbers & other special chars in between the tokens
    if not w.is_stop and w.text not in unwanted_chars and w.text.isalpha():
        tokens.append(w.text)


print(tokens[:200])
```

```python
print(tokens[:500]) #just making sure
```

```python
# getting the word frequencies
word_frequencies = {}
for t in tokens:
    if t not in word_frequencies:
        word_frequencies[t] = 1
    else:
        word_frequencies[t] += 1
```

```python
maximum_frequency = max(word_frequencies.values()) # calculating the maximum frequency

for wrd in word_frequencies:
    word_frequencies[wrd] = word_frequencies[wrd] / maximum_frequency # normalizing the word frequencies


print(word_frequencies)
```

```python
# Tokenizing the sentences
tkn_sentences = [l.text.lower() for l in doc.sents] # making it lower case for my convinence
print(tkn_sentences[:2]) #checking
```

```python
print(tkn_sentences[2])
```

```python
# calculating the sentence scores
sentence_scores = {}
for s in tkn_sentences:
    for word in s.split():  # getting list of words in the sentence
        if word in word_frequencies:  # checking for the word in the frequency dictionary & updating the frequency
            if s not in sentence_scores:
                sentence_scores[s] = word_frequencies[word]
            else:
                sentence_scores[s] += word_frequencies[word]


```

```python
print(sentence_scores)
```

```python
# summarizing the sentences using nlargest module
summarized_sentences = nlargest(5, sentence_scores, key=sentence_scores.get)
```

```python
# back to strings using join
summary = ' '.join([str(snt) for snt in summarized_sentences])
```

```python
# printing the summary
print("Summary:", summary)
print(summary)
```

```python
# calculating lengths
summary_length = len(summary.split())
original_length = len(text.split())

print("Summary length:", summary_length)
print("Original text length:",original_length)
```

```python
from nltk.sentiment import SentimentIntensityAnalyzer
import nltk

nltk.download('vader_lexicon') # downloading VADER sentiment analyzer
```

```python
sia = SentimentIntensityAnalyzer()  # creating a sentiment analyzer instance
sen_srs = sia.polarity_scores(text)

compound = sen_srs['compound'] # getting the sentiment scores
negative = sen_srs['neg']
neutral = sen_srs['neu']
positive = sen_srs['pos']

print(f"Compound: {compound}")
print(f"Negative: {negative}")
print(f"Neutral: {neutral}")
print(f"Positive: {positive}")



```
